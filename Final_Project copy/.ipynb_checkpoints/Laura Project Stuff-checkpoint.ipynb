```python
#df2 = spark.read.csv(SparkFiles.get("ydad3.csv"), sep=",", header=True)
df2 = spark.read.csv("yodad3.csv", sep=",", header=True)
```

```python
tokenizer = Tokenizer(inputCol="Yoda", outputCol="tokens")
```

```python
tokenized = tokenizer.transform(df2)

tokenized['tokens']
```

```python
remover = StopWordsRemover(inputCol="tokens", outputCol="filtered")
```

```python
rmv = remover.transform(tokenized)
```
